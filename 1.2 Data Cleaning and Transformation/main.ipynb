```python
import pandas as pd
```

```python
mobile_df = pd.read_csv('item_details.csv')
mobile_df
```

```python
mobile_df = mobile_df.dropna()
mobile_df
```

```python
mobile_df.info()
```

```python
mobile_df = mobile_df[~mobile_df.apply(lambda row: row.astype(str).str.contains('مطرح نیست')).any(axis=1)]
mobile_df
```

```python
# Use .loc to avoid SettingWithCopyWarning and apply transformations directly
mobile_df.loc[:, 'تعداد سیم‌کارت'] = mobile_df['تعداد سیم‌کارت'].str.replace('۲ عدد', "2")  # Remove ' عدد'
mobile_df.loc[:, 'تعداد سیم‌کارت'] = mobile_df['تعداد سیم‌کارت'].str.replace('۱ عدد', "1")  # Remove ' عدد'
mobile_df.loc[:, 'تعداد سیم‌کارت'] = pd.to_numeric(mobile_df['تعداد سیم‌کارت'])  # Convert to numeric

mobile_df
```

```python
unique_values = mobile_df['حافظهٔ داخلی'].unique()

for i in unique_values:
    print(i)
```

```python
mobile_df.loc[:, 'حافظهٔ داخلی'] = mobile_df['حافظهٔ داخلی'].str.replace('۲۵۶ گیگابایت', "256")  
mobile_df.loc[:, 'حافظهٔ داخلی'] = mobile_df['حافظهٔ داخلی'].str.replace('۳۲ گیگابایت', "32")  
mobile_df.loc[:, 'حافظهٔ داخلی'] = mobile_df['حافظهٔ داخلی'].str.replace('۱۲۸ گیگابایت', "128")  
mobile_df.loc[:, 'حافظهٔ داخلی'] = mobile_df['حافظهٔ داخلی'].str.replace('۸ گیگابایت', "8")  
mobile_df.loc[:, 'حافظهٔ داخلی'] = mobile_df['حافظهٔ داخلی'].str.replace('۵۱۲ گیگابایت', "512")  
mobile_df.loc[:, 'حافظهٔ داخلی'] = mobile_df['حافظهٔ داخلی'].str.replace('۱۶ گیگابایت', "16")  
mobile_df.loc[:, 'حافظهٔ داخلی'] = mobile_df['حافظهٔ داخلی'].str.replace('۶۴ گیگابایت', "64")  
mobile_df.loc[:, 'حافظهٔ داخلی'] = mobile_df['حافظهٔ داخلی'].str.replace('۱ ترابایت', "1000")  
mobile_df.loc[:, 'حافظهٔ داخلی'] = mobile_df['حافظهٔ داخلی'].str.replace('۴ گیگابایت', "4")  
mobile_df.loc[:, 'حافظهٔ داخلی'] = mobile_df['حافظهٔ داخلی'].str.replace('۵۱۲ مگابایت', "0.512") 

mobile_df.loc[:, 'حافظهٔ داخلی'] = pd.to_numeric(mobile_df['حافظهٔ داخلی'])  
mobile_df = mobile_df.rename(columns={'حافظهٔ داخلی': 'حافظه داخلی (گیگابایت)'})

mobile_df
```

```python
unique_values = mobile_df['مقدار رم'].unique()

for i in unique_values:
    print(i)
```

```python
mobile_df.loc[:, 'مقدار رم'] = mobile_df['مقدار رم'].str.replace('۱۲ گیگابایت', "12")  
mobile_df.loc[:, 'مقدار رم'] = mobile_df['مقدار رم'].str.replace('۳ گیگابایت', "3")  
mobile_df.loc[:, 'مقدار رم'] = mobile_df['مقدار رم'].str.replace('۴ گیگابایت', "4")  
mobile_df.loc[:, 'مقدار رم'] = mobile_df['مقدار رم'].str.replace('۸ گیگابایت', "8")  
mobile_df.loc[:, 'مقدار رم'] = mobile_df['مقدار رم'].str.replace('۱ گیگابایت', "1")  
mobile_df.loc[:, 'مقدار رم'] = mobile_df['مقدار رم'].str.replace('۶ گیگابایت', "6")  
mobile_df.loc[:, 'مقدار رم'] = mobile_df['مقدار رم'].str.replace('۲ گیگابایت', "2")  
mobile_df.loc[:, 'مقدار رم'] = mobile_df['مقدار رم'].str.replace('۱.۵ گیگابایت', "1.5")  
mobile_df.loc[:, 'مقدار رم'] = mobile_df['مقدار رم'].str.replace('۵۱۲ مگابایت', "0.512") 


mobile_df.loc[:, 'مقدار رم'] = pd.to_numeric(mobile_df['مقدار رم'])  
mobile_df = mobile_df.rename(columns={'مقدار رم': 'مقدار رم (گیگابایت)'})

mobile_df
```

```python
unique_values = mobile_df['مقدار رم (گیگابایت)'].unique()

for i in unique_values:
    print(i)
```

```python
from num2words import num2words
import re

# Function to convert Persian numerals to English numerals
def convert_persian_to_english(num_str):
    # Remove تومان and any commas
    num_str = num_str.replace(' تومان', '').replace('٬', '')
    
    # Remove any non-numeric characters (including any text in parentheses)
    num_str = re.sub(r'\D', '', num_str)
    
    return int(num_str)  # Convert to integer if needed

# Apply transformations
mobile_df['قیمت'] = mobile_df['قیمت'].apply(convert_persian_to_english)
mobile_df = mobile_df.rename(columns={'قیمت': 'Price(Toman)'})
mobile_df = mobile_df.rename(columns={'رنگ': 'Color'})
mobile_df = mobile_df.rename(columns={'مقدار رم (گیگابایت)': 'RAM(GB)'})
mobile_df = mobile_df.rename(columns={'حافظه داخلی (گیگابایت)': 'Internal Storage(GB)'})
mobile_df = mobile_df.rename(columns={'اصالت برند': 'Brand Origin'})
mobile_df = mobile_df.rename(columns={'تعداد سیم‌کارت': 'SIM Count'})
mobile_df = mobile_df.rename(columns={'وضعیت': 'Status'})
mobile_df = mobile_df.rename(columns={'برند و مدل': 'Brand and Model'})

mobile_df
```

```python
# Function to remove English words using regex
def remove_english_words(text):
    return re.sub(r'[a-zA-Z]+', '', text)

# Apply the function to the 'رنگ' column
mobile_df['Color'] = mobile_df['Color'].apply(remove_english_words)
mobile_df = mobile_df[mobile_df['Color'].str.strip() != ""]
mobile_df
```
